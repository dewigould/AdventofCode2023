## Part 2
- Wow, this was insanely difficult. But I did it.
- It would be way too computationally intense to pipe through all the seeds to find out their locations and brute force it
- Instead you have to basically work backwards.
- I start from all the ranges of possible locations (i.e. the ranges of the final map)
- from there, you look at the domains of the final map which correpond to this
- I built an "overlap" function, so I find the overlap of the domain of the final map with the range of the penultimate map
- This means I don't need to consider EVERY SINGLE seed going through the final map, only the smallest value of each interval in this list of overlaps.
- I then use the mapping to pipe this list of ranges BACKWARDS through the 5th map to get a new set of domains for the fifth map
- I then iteratively continue this through all of the maps: first overlapping range and domains when they are next to eachtoher, then mapping backwards through
- Once I finally end up with a list of ranges of seed numbers, I only have to pipe through the lowest values in each range.
- Basically the final result I have is that I've combined all the maps into one large map, meaning I need only consider the lowest values of each pair

```python
seed_ranges = []
```

```python
# Parse the seed ranges that we have as input
for i in range(int(len(seed_nums)/2)):
    seed_ranges.append((seed_nums[2*i],seed_nums[2*i]+seed_nums[2*i+1]))
seed_ranges
    
```

```python
class Pipeline:
    """
    Class representing one of the maps
    list_of_maps is a list of tuples representing domain:range 
    domain/ range are self-explanatory, but they are padded out with (0,X) and (Y,10^25) as boundaries
    backwards = True/False is a parameter that lets me use this pipeline to either map forwards or backwards
    .pipe is a method which takes in a number and gives the output
    """

    def __init__(self,list_of_maps,backwards):
        
        self.list_of_maps = sorted(list_of_maps, key=lambda x: x[0])

        self.domain = [i[0] for i in list_of_maps]
        self.domain.sort()
        if 0 not in [i[0] for i in self.domain]:
            min_val = min([i[0] for i in self.domain])
            self.domain = [(0,min_val-1)]+self.domain
            
        max_val = max([i[1] for i in self.domain])
        self.domain = self.domain + [(max_val+1,10**25)]
        
        
        
        self.range = [i[1] for i in list_of_maps]

        if 0 not in [i[0] for i in self.range]:
            min_val = min([i[0] for i in self.range])
            self.range = [(0,min_val-1)]+self.range
            
        max_val = max([i[1] for i in self.range])
        self.range = self.range + [(max_val+1,10**25)]
        
        
        
        if backwards:
            self.list_of_maps = [(i[1],i[0]) for i in self.list_of_maps]
            d = self.domain.copy()
            r = self.range.copy()
            self.domain = r
            self.range = d
        
        
    def pipe(self,input_val):
        for j in self.list_of_maps:
            if input_val >=j[0][0] and input_val<=j[0][1]:
                return j[1][0]+(input_val-j[0][0])
        
        return input_val
        

        
```

```python
# Find the list of overlaps of two intervals (x,y) and (a,b)
def find_overlap(interval1, interval2):
    start1, end1 = interval1
    start2, end2 = interval2

    # Check for no overlap
    if end1 < start2 or end2 < start1:
        return None

    # Calculate overlap
    overlap_start = max(start1, start2)
    overlap_end = min(end1, end2)

    return overlap_start, overlap_end
        
```

```python
# Find overlaps of list of ranges and domains
def get_new_range_based_on_overlap(range_input,target_domain):
    new_range_1 = []
    for r in range_input:
        for d in target_domain:
            overlap = find_overlap(r,d)
            if overlap != None:
                new_range_1.append(overlap)
    return(new_range_1)

```

```python
def pipe_backwards(range_pair,model):
    return(model.pipe(range_pair[0]),model.pipe(range_pair[1]))
```

```python
# This merges the pipelines!
pipes = [Pipeline(mapping_instructions[i],False) for i in np.arange(1,8)]
backward_pipes = [Pipeline(mapping_instructions[i],True) for i in np.arange(1,8)]

#This is the main algo, we start with the domain of the final map
current_domain = pipes[6].domain
#Then we final overlaps with the range of the map before it
next_step=get_new_range_based_on_overlap(current_domain,pipes[5].range)
#Then we pipe these overlaps back through the penultimate map to land of domains 
next_step = [pipe_backwards(i,backward_pipes[5]) for i in next_step]
# Then we overlap these domains with the range of the map before this one...etc.
next_step=get_new_range_based_on_overlap(next_step,pipes[4].range)
next_step = [pipe_backwards(i,backward_pipes[4]) for i in next_step]
next_step=get_new_range_based_on_overlap(next_step,pipes[3].range)
next_step = [pipe_backwards(i,backward_pipes[3]) for i in next_step]
next_step=get_new_range_based_on_overlap(next_step,pipes[2].range)
next_step = [pipe_backwards(i,backward_pipes[2]) for i in next_step]
next_step=get_new_range_based_on_overlap(next_step,pipes[1].range)
next_step = [pipe_backwards(i,backward_pipes[1]) for i in next_step]
next_step=get_new_range_based_on_overlap(next_step,pipes[0].range)
next_step = [pipe_backwards(i,backward_pipes[0]) for i in next_step]
next_step
```

```python
pipeline_of_pipes = [Pipeline(mapping_instructions[i],False) for i in np.arange(1,8)]
```

```python
# Combine all pipes to get one final number
def pipe_through_whole_thing(x):
    output = x
    for j in np.arange(0,7):        
        output = pipeline_of_pipes[j].pipe(output)
    return(output)
```

```python
# Output of algo above is a list of domains
# Need to overlap this with the input seed ranges, but being careful to do a more sophisticated overlap 
# because we have prescribed list of input seeds, it's not just the generic overlap!
def find_overlap_with_seeds(interval_a, interval_b):

    start_a, end_a = interval_a
    start_b, end_b = interval_b

    # Check for no overlap
    if end_a < start_b or end_b < start_a:
        return None

    # Calculate overlap
    overlap_start = max(start_a, start_b)
    overlap_end = min(end_a, end_b)

    # Ensure overlap elements are in interval_a
    overlap_start = max(overlap_start, start_a)
    overlap_end = min(overlap_end, end_a)

    # Check if there is still an overlap after adjusting for interval_a
    if overlap_end < overlap_start:
        return None

    return overlap_start, overlap_end

```

```python
# Compute final seed overlap
final_seeds = []
for s in seed_ranges:
    for n in next_step:
        output = find_overlap_with_seeds(s,n)
        if output!=None:
            final_seeds.append(output)
```

```python
# Final the seed values at the beginning of each of the ranges
min_vals = [i[0] for i in final_seeds]
# Pipe these seed values through the pipeline.
results = [pipe_through_whole_thing(i) for i in min_vals]
```
